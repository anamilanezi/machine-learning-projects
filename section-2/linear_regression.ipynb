```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv("cost_revenue_clean.csv")
```

```python
data.describe()
```

```python
3.29e7
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
x_array = data['production_budget_usd']
```

```python
# Create a scatter plot using data extracted from csv
plt.figure(figsize=(10,6)) 
plt.scatter(X, y, alpha=0.3)

# Add a title 
plt.title('Film Const vs Global Revenue')

# Add label to x and y axis:
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')

# Changing the x and y range
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)


plt.show()
```

```python
# Create an object instance of LinearRegression class
regression = LinearRegression()
regression.fit(X, y)
```

```python
# Slope Coefficient (coeficiente angular) theta_1
regression.coef_
```

```python
# Intercept (interseco)
regression.intercept_
```

* The slope coefficient 3.1 shows a positive relation between budget and revenue (for each USD1 invested -> USD3 revenue) 
* The intercept shows that a movie with a budget of 0 would lose 7 million, which it's not realistic.  

* Our model is a simplification of the real world. If we wanted to know how much revenue a movie with 500000000 of investiment would gain:
h0 = -7236192 + 3.11x -> h0 = -7236192 + 3.11 * 50000000

```python
# Create a scatter plot using data extracted from csv
plt.figure(figsize=(10,6)) 
plt.scatter(X, y, alpha=0.3)

# Plot a line on the chart: first value won't work with a df
plt.plot(x_array, regression.predict(X), color='red', linewidth=4)

# Add a title 
plt.title('Film Const vs Global Revenue')

# Add label to x and y axis:
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')

# Changing the x and y range
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)


plt.show()
```

R square the amount of variation in film revenue that is explained by the film's budget

```python
# Calculating the R square
regression.score(X, y)
```
